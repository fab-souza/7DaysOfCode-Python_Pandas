# Variáveis categóricas:

## Tipo de usuário:

```python
dados.tipo_vinculo_usuario.value_counts()
```

```python
def mostrar_frequencia(variavel):
  frequencia = dados[variavel].value_counts()
  return frequencia
```

```python
mostrar_frequencia('tipo_vinculo_usuario')
```



---



```python
dados.tipo_vinculo_usuario.value_counts(normalize = True).round(4) * 100
```

```python
def mostrar_percentual(variavel):
  percentual = dados[variavel].value_counts(normalize = True).round(4) * 100
  return percentual
```

```python
mostrar_percentual('tipo_vinculo_usuario')
```

```python
def tabela_freq_porcent(variavel):
  frequencia = dados[variavel].value_counts()
  percentual = dados[variavel].value_counts(normalize = True).round(4) * 100
  tabela = pd.DataFrame({'Frequência' : frequencia, 'Porcentagem (%)' : percentual})
  return tabela
```

```python
tabela_freq_porcent('tipo_vinculo_usuario')
```

```python
def tabela_crosstab_freq(variavel2):
  tabela = pd.crosstab(dados[variavel2], dados.tipo_vinculo_usuario)
  return tabela
```
