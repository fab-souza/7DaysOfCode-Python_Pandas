# Calculando percentual

## Criando uma função:

```python
def calculo_percentual(A, B):
  resultado = round((B * 100 / A) - 100, 2)
  return resultado
```

```python
calculo_percentual(2538, 2316)
```

```python
# 2017
cursos_e_anos_pivot.iloc[:,0]
```

```python
# 2018
cursos_e_anos_pivot.iloc[:,1]
```

```python
calculo_percentual(cursos_e_anos_pivot.iloc[:,0], cursos_e_anos_pivot.iloc[:,1])
```

# Criando a tabela


```python
diferenca_2017_2018 = calculo_percentual(cursos_e_anos_pivot.iloc[:,0], cursos_e_anos_pivot.iloc[:,1])
diferenca_2018_2019 = calculo_percentual(cursos_e_anos_pivot.iloc[:,1], cursos_e_anos_pivot.iloc[:,2])
diferenca_2019_2022 = calculo_percentual(cursos_e_anos_pivot.iloc[:,2], cursos_e_anos_pivot.iloc[:,3])
```

```python
diferencas_percentuais = pd.DataFrame({'2017-2018' : diferenca_2017_2018,
                                       '2018-2019' : diferenca_2018_2019,
                                       '2019-2022' : diferenca_2019_2022})
diferencas_percentuais
```

```python
diferencas_percentuais.columns
```

```python
diferencas_percentuais.reset_index(inplace = True)
diferencas_percentuais.Curso = diferencas_percentuais.Curso.str.capitalize()
diferencas_percentuais
```
