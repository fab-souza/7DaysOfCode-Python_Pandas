# Empréstimos de 2015 a 2020

```python
emprestimos.query("tipo_vinculo_usuario == 'ALUNO DE GRADUAÇÃO'")
```

```python
emprestimos_graduacao = emprestimos.query("tipo_vinculo_usuario == 'ALUNO DE GRADUAÇÃO'")
emprestimos_graduacao.query('data_emprestimo > 2015')
```

```python
emprestimos_graduacao = emprestimos_graduacao.query('data_emprestimo > 2015')
emprestimos_graduacao.reset_index(drop = True, inplace = True)
emprestimos_graduacao.head(3)
```

## Selecionando os cursos

```python
alunos.curso.value_counts()
```

```python
alunos_selecionados = alunos.query("curso == ['BIBLIOTECONOMIA','CIÊNCIAS SOCIAIS','COMUNICAÇÃO SOCIAL','DIREITO','FILOSOFIA','PEDAGOGIA']")
alunos_selecionados
```

## Merge

```python
emprestimos_graduacao.loc[:,['matricula_ou_siape','data_emprestimo']]
```

```python
emprestimos_graduacao = emprestimos_graduacao.loc[:,['matricula_ou_siape','data_emprestimo']]
cursos_e_anos_selecionados = emprestimos_graduacao.merge(alunos_selecionados)
cursos_e_anos_selecionados.head(3)
```

```python
cursos_e_anos_selecionados.data_emprestimo.dt.year
```

```python
cursos_e_anos_selecionados.data_emprestimo = cursos_e_anos_selecionados.data_emprestimo.dt.year
cursos_e_anos_selecionados.iloc[:,[1,3]]
```

```python
cursos_e_anos_selecionados.iloc[:,[1,3]].value_counts()
```

```python
cursos_e_anos_selecionados.iloc[:,[1,3]].value_counts().reset_index()
```

```python
cursos_e_anos = cursos_e_anos_selecionados.iloc[:,[1,3]].value_counts().reset_index()
cursos_e_anos.columns = ['Ano','Curso','Emprestimos']
cursos_e_anos
```

# pivot_table

```python
emprestimos_curso_anos = cursos_e_anos.pivot_table(
        index = 'Curso',
        columns = 'Ano',
        values = 'Emprestimos',
        fill_value = '-',
        aggfunc= sum,
        margins = True,
        margins_name = 'Total',
)
emprestimos_curso_anos
```
