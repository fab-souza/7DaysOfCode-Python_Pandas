# Empréstimos ao longo do tempo:

```python
dados.data_emprestimo.dt.date
```

```python
dados.data_emprestimo.dt.date.value_counts()
```

```python
dias_emprestimo = pd.DataFrame(data = dados.data_emprestimo.dt.date.value_counts())
dias_emprestimo
```

```python
dias_emprestimo.plot()
plt.title('Empréstimos da UFRN')
plt.show();
```

```python
dias_emprestimo.reset_index(inplace = True)
dias_emprestimo.rename(columns = {'index' : 'Dia', 'data_emprestimo' : 'Quantidade_emprestimo'}, inplace = True)
dias_emprestimo
```

```python
fig = plt.subplots(figsize = (11, 5))
sns.lineplot(data = dias_emprestimo, x = 'Dia', y = 'Quantidade_emprestimo').set(title = 'Empréstimo por Dia');
```

# Empréstimo por mês:

```python
dados.data_emprestimo.dt.month
```

```python
dados.data_emprestimo.dt.month.value_counts()
```

```python
mes_emprestimo = pd.DataFrame(data = dados.data_emprestimo.dt.month.value_counts())
mes_emprestimo.sort_index(inplace = True)
mes_emprestimo.index.name = 'Mes'
mes_emprestimo.reset_index(inplace = True)
mes_emprestimo.rename(columns = {'data_emprestimo' : 'Quantidade_emprestimo'}, inplace = True)
mes_emprestimo
```

```python
meses = {
    1 : 'Janeiro',
    2 : 'Fevereiro',
    3 : 'Marco',
    4 : 'Abril',
    5 : 'Maio',
    6 : 'Junho',
    7 : 'Julho',
    8 : 'Agosto',
    9 : 'Setembro',
    10 : 'Outubro',
    11 : 'Novembro',
    12 : 'Dezembro'
}
```

```python
mes_emprestimo.Mes = mes_emprestimo.Mes.map(meses)
mes_emprestimo
```

```python
fig = plt.subplots(figsize = (11, 5))
sns.lineplot(data = mes_emprestimo, x = 'Mes', y = 'Quantidade_emprestimo').set(title = 'Empréstimo por Mês');
```

# Empréstimo por hora:

```python
dados.data_emprestimo.dt.hour
```

```python
dados.data_emprestimo.dt.hour.value_counts()
```

```python
hora_emprestimo = pd.DataFrame(data = dados.data_emprestimo.dt.hour.value_counts())
hora_emprestimo.reset_index(inplace = True)
hora_emprestimo.rename(columns = {'index' : 'Hora', 'data_emprestimo' : 'Quant_emprestimo'}, inplace = True)
hora_emprestimo
```

```python
ax = plt.subplots(figsize = (11, 5))
sns.barplot(data = hora_emprestimo, x = 'Hora', y = 'Quant_emprestimo', palette = 'Blues', 
            hue = 'Quant_emprestimo', dodge = False).set(title = 'Empréstimo por Hora')
plt.legend([],[], frameon = False)
ax = ax;
```
